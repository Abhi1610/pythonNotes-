```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot
import seaborn as sns
%matplotlib inline
```

```python
tips =sns.load_dataset('tips')
```

```python
tips.head(20)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
#hue ->in seaborn the hue parameter represents which column in the dataframe you want to use for color encoding.
sns.boxplot(x='day',y='total_bill',hue="smoker",data=tips)
```

```python
tips['day'].count()
```

```python
sns.countplot(x='day',data=tips,hue="smoker")
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue="sex")
```

```python
tips['total_bill'].min()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
sns.swarmplot(x='sex',y='total_bill',data=tips,hue="smoker")
```

```python
sns.swarmplot(x='time',y='tip',data=tips,order=['Dinner','Lunch'],size=7)
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette="coolwarm")
```

```python
tips.corr()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',row='sex',col='time',data=tips)
```

```python
sns.heatmap(tips.corr(),annot=True)
```
